```python
"""""
Sean Cummins 9/07/19
Playing with aotools phase screen generation classes with typical turbulence parameter values. 
"""""

import aotools
import numpy
import matplotlib.pyplot as plt

r0 = 0.1 #Fried parameter [m] - typically 5-10 cm for vertical viewing 
N = 100 #pixel number
delta = 0.02 #length scale of pixels [m] - 2 m screen
L0 = 100.0 #Length of outer scale [m] - typically 100 m
l0 = 0.01 #Length of inner scale [m] - typically 1-100 mm

ps = aotools.turbulence.phasescreen.ft_phase_screen(r0, N, delta, L0, l0, FFT=None, seed=None) #[rad] 

plt.imshow(ps)
plt.colorbar()
```
